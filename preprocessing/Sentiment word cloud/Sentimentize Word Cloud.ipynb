```python
with open('word_counts.json', 'r') as f:
    word_counts = json.load(f)
```

```python
with open('top_25_subreddit_names.json', 'r') as f:
    subreddits = json.load(f)
```

```python
polarity = defaultdict(lambda: defaultdict(int))
total = defaultdict(lambda: defaultdict(int))
```

```python
def sentimentize_data(data):
    global polarity
    global total
    for comment in data:
        if comment['subreddit'] in subreddits: 
            subreddit_word_counts = word_counts[comment['subreddit']]
            comment_lowercase = comment['body'].lower()
            for dct in subreddit_word_counts:
                if dct['key'] in comment_lowercase:
                    polarity[comment['subreddit']][dct['key']] += TextBlob(comment['body']).sentiment.polarity
                    total[comment['subreddit']][dct['key']] += 1
```

```python
def main():
    global output
    output = []
    rootdir = '/Users/ashackelford/Desktop/Reddit/'
    for subdir, dirs, files in os.walk(rootdir):
        for file in files:
            with open(subdir + "/" + file, 'r') as f:
                print(file)
                start_time = time.time()
                sentimentize_data(json.load(f))
                print (time.time() - start_time)
```

```python
main()
```

```python
word_counts_results = {}
for subreddit, values in word_counts.items():
    word_counts_results[subreddit] = []
    for dct in values:
        out_dct = {}
        out_dct['key'] = dct['key']
        out_dct['value'] = dct['value']
        out_dct['sentiment'] = polarity[subreddit][dct['key']] / total[subreddit][dct['key']]
        word_counts_results[subreddit].append(out_dct)
```
