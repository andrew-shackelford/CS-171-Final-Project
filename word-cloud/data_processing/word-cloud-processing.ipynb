```python
from textblob import TextBlob
import json
import pickle
import os
import sys
import string 
from nltk.corpus import words
from nltk.corpus import stopwords

from collections import defaultdict
```

```python
import enchant
d = enchant.Dict("en_US")
```

```python
word_count = []  
stop_words = set(stopwords.words('english')) 

with open('/Users/jenniferlee/Documents/cs171/project/word-cloud/data/top100wordsPerTop25.json') as json_file:
    data = json.load(json_file)
    for subreddit in data:
        temp = subreddit["words"].copy()
        for word in subreddit["words"]:
            if d.check(word[0]) == False or len(word[0]) == 1 or word[0] == "..." or word[0] in stop_words:
                temp.remove(word) 
                subreddit["words"] = temp
    
    for subreddit in data:
        subreddit["words-only"] = [word[0] for word in subreddit["words"]]
        subreddit["counts-only"] = [word[1] for word in subreddit["words"]]

    newDict = {}
    for subreddit in data:
        newDict[subreddit["key"]] = dict(zip(subreddit["words-only"], subreddit["counts-only"]))
        
    newJson = {}
    for subreddit in data:
        newJson[subreddit["key"]] = []
        for word in subreddit["words"]:
            newJson[subreddit["key"]].append({"key":word[0],"value":word[1]})
```

```python
with open('/Users/jenniferlee/Documents/cs171/project/word-cloud/data/top100wordsPerTop25Filtered.json', 'w') as outfile:
    json.dump(data, outfile, indent=4)
```

```python
with open('/Users/jenniferlee/Documents/cs171/project/word-cloud/data/wordCounts.json', 'w') as outfile:
    json.dump(newDict, outfile, indent=4)
```

```python
with open('/Users/jenniferlee/Documents/cs171/project/word-cloud/data/wordCountsAsJSON.json', 'w') as outfile:
    json.dump(newJson, outfile, indent=4)
```
